```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("station_data/data_station1.csv")

dt_index = pd.date_range( start="2010-01-01 00:00:00", end="2023-03-31 23:00:00", freq="H")

df['datetime'] = dt_index
df = df.set_index("datetime")
plt.figure(figsize=(12,6))
plt.plot(dt_index, df['co'])

plt.title("Time vs CO Concentration")
plt.xlabel("Time")
plt.ylabel("CO concentration")
plt.show()
```

```python
data = df.loc["2020-01-01":"2023-3-1"]
```

```python
import numpy as np
data["t"] = np.arange(len(data))
data['cos_year'] = np.cos(2*np.pi*data['t']/ (365*24))
data['sin_year'] = np.sin(2*np.pi*data['t']/ (365*24))

data['cos_day'] = np.cos(2*np.pi*data['t']/24)
data['sin_day'] = np.sin(2*np.pi*data['t']/24)
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt

x_vars = ['t', 'cos_year', 'sin_year', 'cos_day', 'sin_day']

plt.figure(figsize=(12,15))

for i, var in enumerate(x_vars, 1):
    plt.subplot(5,1,i)
    plt.scatter(data[var], data['co'], s=1)
    plt.xlabel(var)
    plt.ylabel("CO")
    plt.title(f"CO vs {var}")

plt.tight_layout()
plt.show()
```

```python
import statsmodels.api as sm
cols = ['co','cos_year','sin_year','cos_day','sin_day']
clean = data[cols].dropna()

X = clean[['cos_year','sin_year','cos_day','sin_day']]
X = sm.add_constant(X)

y = clean['co']

model = sm.OLS(y, X).fit()

print(model.summary())
print(model.params)
```

```python
clean['trend'] = model.predict(X)
clean['residual'] = clean['co'] - clean['trend']
```

```python
from statsmodels.tsa.arima.model import ARIMA

arma_model = ARIMA(clean['residual'], order=(1,0,1))
arma_fit = arma_model.fit()

print(arma_fit.summary())
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(arma_fit.resid, lags=40)
plt.show()
```

```python
arma_model2 = ARIMA(clean['residual'], order=(2,0,1))
arma_fit2 = arma_model2.fit()

print(arma_fit2.summary())
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(arma_fit2.resid, lags=40)
plt.show()
```

```python
arma_model3 = ARIMA(clean['residual'], order=(3,0,1))
arma_fit3 = arma_model3.fit()

print(arma_fit3.summary())
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(arma_fit3.resid, lags=40)
plt.show()
```

```python
residual_forecast = arma_fit.forecast(steps=168)
future_t = np.arange(len(clean), len(clean)+168)

cos_year = np.cos(2*np.pi*future_t/(365*24))
sin_year = np.sin(2*np.pi*future_t/(365*24))

cos_day = np.cos(2*np.pi*future_t/24)
sin_day = np.sin(2*np.pi*future_t/24)

future_X = np.column_stack([cos_year,sin_year,cos_day,sin_day])
future_X = sm.add_constant(future_X)

seasonal_forecast = model.predict(future_X)
final_forecast = seasonal_forecast + residual_forecast
```

```python
clean['day_of_year'] = clean.index.dayofyear
clean['hour'] = clean.index.hour
X = pd.get_dummies(clean[['day_of_year', 'hour']], 
                   columns=['day_of_year', 'hour'], 
                   drop_first=True)

import statsmodels.api as sm
X = sm.add_constant(X)

# 🔥 FIX
X = X.astype(float)

model = sm.OLS(clean['co'], X).fit()

clean['residual'] = clean['co'] - model.predict(X)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(clean['residual'], lags=40)
plot_pacf(clean['residual'], lags=40)
```

```python
from sklearn.metrics import r2_score

# Combined prediction = seasonal trend + ARMA fitted values
clean['final_fitted'] = clean['trend'] + clean['arma_fitted']

r2 = r2_score(clean['co'], clean['final_fitted'])
print(f"Combined Model R² Score: {r2:.4f}")
```

```python
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import r2_score
import numpy as np

split = int(len(clean) * 0.8)
train = clean.iloc[:split]
test = clean.iloc[split:]

X_train = train[['cos_year', 'sin_year', 'cos_day', 'sin_day']].dropna()
X_train = sm.add_constant(X_train)
y_train = train['co'].dropna()
ols_model = sm.OLS(y_train, X_train).fit()

train_trend = ols_model.predict(X_train)
train_residual = y_train - train_trend
arma_fit = ARIMA(train_residual, order=(1,0,1)).fit()

X_test = test[['cos_year', 'sin_year', 'cos_day', 'sin_day']].dropna()
X_test = sm.add_constant(X_test)
seasonal_pred = ols_model.predict(X_test)
y_test = test['co'].dropna()
actual_test_residual = y_test.values - seasonal_pred.values

full_residual = np.concatenate([train_residual.values, actual_test_residual])
arma_refit = ARIMA(full_residual, order=(1,0,1)).fit()

residual_pred = arma_refit.fittedvalues[len(train_residual):]
final_pred = seasonal_pred.values + residual_pred

train_r2 = r2_score(y_train, train_trend + arma_fit.fittedvalues)
test_r2 = r2_score(y_test, final_pred)

print(f"Train R²: {train_r2:.4f}")
print(f"Test R²:  {test_r2:.4f}")
```
